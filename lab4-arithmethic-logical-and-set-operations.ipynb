```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

# **4.1 Arithetic Operations**

# **4.1.1 Addition**

```python
img1= cv2.imread('/kaggle/input/datasets/abhashkharel7/imageippr/einstein.jpg')
img2= cv2.imread('/kaggle/input/datasets/abhashkharel7/imageippr/flower.jpg')
img2= cv2.resize(img2,(img1.shape[1],img1.shape[0]))
added_image= cv2.add(img1,img2)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title(' Image 2')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(added_image, cmap='gray')
plt.title('Added Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.1.2 Subtraction**

```python
subtracted_image = cv2.subtract(img1, img2)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title(' Image 2')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(subtracted_image, cmap='gray')
plt.title('Subtracted Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.1.3 Multiplication by scalar(k)**

```python
scalar = 1.5
img1_float = img1.astype(np.float32)
scaled_img = img1_float * scalar
scaled_img = scaled_img.astype(np.uint8)
scaled_img = np.clip(scaled_img, 0, 255)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(scaled_img, cmap='gray')
plt.title('Scaled Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.1.4 Multiplication of two images**

```python
img1_float = img1.astype(np.float32) / 255
img2_float = img2.astype(np.float32) / 255
multiplied_img = img1_float * img2_float
multiplied_img = (multiplied_img * 255).astype(np.uint8)
multiplied_img = np.clip(multiplied_img, 0, 255)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Image 1')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title(' Image 2')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(multiplied_img, cmap='gray')
plt.title(' Multiplied Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.2 Logical Operations**

# **4.1.2 AND**

```python
img3= np.zeros((200,200),dtype=np.uint8)
img4= np.zeros((200,200),dtype=np.uint8)
cv2.rectangle(img3,(30,30),(130,130),255,-1)
cv2.rectangle(img4,(80,80),(180,180),255,-1)
and_result= cv2.bitwise_and(img3,img4)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(img3, cmap='gray')
plt.title('Image ')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img4, cmap='gray')
plt.title(' Image ')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(and_result, cmap='gray')
plt.title(' And ')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.2.2 OR**

```python
or_result= cv2.bitwise_or(img3,img4)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(img3, cmap='gray')
plt.title('Image ')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img4, cmap='gray')
plt.title(' Image ')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(or_result, cmap='gray')
plt.title(' Or ')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.2.3 NOT**

```python
not_result= cv2.bitwise_not(img3)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(img3, cmap='gray')
plt.title('Image 1')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(not_result, cmap='gray')
plt.title('Not')
plt.axis('off')
plt.tight_layout()
plt.show()
```

# **4.2.4 X-OR**

```python
xor_result= cv2.bitwise_xor(img3,img4)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.imshow(img3, cmap='gray')
plt.title('Image ')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img4, cmap='gray')
plt.title(' Image ')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(xor_result, cmap='gray')
plt.title(' X-or ')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/kaggle/input/datasets/abhashkharel7/imageippr/bellingham.jpeg', cv2.IMREAD_GRAYSCALE).astype(np.float32)
img
```

# **Gaussian Noise**

```python
mean =0
std = 50

gaussian_noise = np.random.normal(mean , std , img.shape)
noise_mask = cv2.normalize(gaussian_noise , None , 0,255,cv2.NORM_MINMAX ).astype(np.uint8)

noisy_img = np.clip(img + gaussian_noise , 0, 255).astype(np.uint8)


plt.figure(figsize = (20,5))

plt.subplot(1,4,1)
plt.imshow(img , cmap ='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(noise_mask , cmap ='gray')
plt.title("Noise Mask")
plt.axis('off')

plt.subplot(1,4,3)
plt.hist(gaussian_noise.ravel(),bins= 100 , color ='black')
plt.title('Noise Histogram')

plt.subplot(1,4,4)
plt.imshow(noisy_img , cmap ='gray')
plt.title("Noise Image")
plt.axis('off')
```

# **Rayleigh Noise**

```python
scale = 80

rayleigh_noise = np.random.rayleigh(scale , img.shape)
noise_mask = cv2.normalize(rayleigh_noise , None , 0,255,cv2.NORM_MINMAX ).astype(np.uint8)

noisy_img = np.clip(img + rayleigh_noise , 0, 255).astype(np.uint8)


plt.figure(figsize = (20,5))

plt.subplot(1,4,1)
plt.imshow(img , cmap ='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(noise_mask , cmap ='gray')
plt.title("Rayleigh Noise Mask")
plt.axis('off')

plt.subplot(1,4,3)
plt.hist(rayleigh_noise.ravel(),bins= 100 , color ='black')
plt.title('Noise Histogram')

plt.subplot(1,4,4)
plt.imshow(noisy_img , cmap ='gray')
plt.title("Noise Image")
plt.axis('off')
```
